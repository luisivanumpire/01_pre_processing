### choose some countries with area less than 50 km2 and poblation grather than 500

```python
countries[ (countries["km2"]<50) & (countries["poblacion_miles"]>500) ]
```

### choose some countries with area less than 5 km2 or poblation less than 5

```python
countries[(countries["km2"]<5) | (countries["poblacion_miles"]<5)]
```

### Choose tiny countries (less than 50 km2) with a little poblation (less than 50) and belong to Europe

```python
countries[(countries["continente"]=="Europa") & 
          (countries["km2"]<50) & 
          (countries["poblacion_miles"]<50)]
```

### Choose tiny countries (less than 50 km2) with a little poblation (less than 50) and don't belong to Europe

```python
countries[(countries["continente"]!="Europa") & 
          (countries["km2"]<50) & 
          (countries["poblacion_miles"]<50)]
```

```python
countries.loc[(countries["continente"]!="Europa") & 
          (countries["km2"]<50) & 
          (countries["poblacion_miles"]<50)]
```
