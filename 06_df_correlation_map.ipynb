```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv('personas.txt')

people = pd.DataFrame(data)
people
```

### The relation between high (altura) and weight (peso)

```python
plt.scatter(people["altura"], people["peso"])
plt.xlabel("High (cm)")
plt.ylabel("Weight (kg)")
plt.show()
```

There is a lineal relation

### Relation between income and working hours

```python
plt.scatter(people["horas_trabajadas"], people["ingreso"])
plt.xlabel("Working hours (Hr)")
plt.ylabel("Income ($)")
plt.show()
```

### Relation between income and absence

```python
plt.scatter(people["ausencias"], people["ingreso"])
plt.xlabel("Absence (days)")
plt.ylabel("Income ($)")
plt.show()
```

There is a negative relation. The graph is a little rare, because we are using a continuous var and a discrete var.

### Relation between income and weight

```python
plt.scatter(people["peso"], people["ingreso"])
plt.xlabel("Weight (Kg)")
plt.ylabel("Income ($)")
plt.show()
```

There are not relation between income and weight

### Correlation matrix

```python
people.corr()
```

```python
corr_matrix = people.corr()
plt.matshow(corr_matrix, cmap="bwr", vmin=-1, vmax=1) # bwr -1 0 +1
plt.xticks(range(5), people.columns, rotation=90)
plt.yticks(range(5), people.columns)
plt.colorbar()
plt.show()
```

```python
corr_matrix = people.corr()
plt.matshow(corr_matrix, cmap="bwr", vmin=-1, vmax=1) # bwr -1 0 +1
plt.xticks(range(5), people.columns, rotation=90)
plt.yticks(range(5), people.columns)

for i in range(len(corr_matrix.columns)):
    for j in range(len(corr_matrix.columns)):
        plt.text(i,j,round(corr_matrix.iloc[i,j],2),
                ha="center", va="center")

plt.colorbar()
plt.show()
```
